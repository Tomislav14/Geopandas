```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python

# Putanja shapefilea
shapefile_path = 'E:/Otvoreni GIS podaci Grada Zagreba/Telekomunikacijski_razvodni_ormarici.shp'
# Uitaj shapefile using GeoPandas
data = gpd.read_file(shapefile_path, encoding='latin1')  # Change the encoding if necessary
# Plotaj mapu
data.plot(figsize=(15, 8), color='pink', edgecolor='black')
# Dodaj naslov
plt.title('Telekomunikacijski razvodni ormarii')
# Prikai kartu
plt.show()
```
